```python
from mindquantum.device import LinearQubits, GridQubits
topo=GridQubits(2,5)
from mindquantum.algorithm.nisq import MaxCutAnsatz
from mindquantum.io.display import draw_topology
from IPython.display import display_svg
display_svg(draw_topology(topo))
```

```python
g=[(0,1),(1,2),(2,3),(3,4),(0,5),(1,6),(2,7),(3,8),(4,9),(5,6),(6,7),(7,8),(8,9)]


maxcut = MaxCutAnsatz(g, 1) # QAOA depth 1
circ=maxcut.circuit
# print('origin',circ)
circ.svg()
```

```python
#begin sabre

from mindquantum.device import SABRE

solver = SABRE(circ, topo)
new_circ, init_mapping, final_mapping = solver.solve(5, 0.5, 0.3, 0.2)
print('Init:',init_mapping,'\nFinal:',final_mapping)
print('gates in circuit:', len(circ))
print('  in new circuit:', len(new_circ))
# print('new:\n',new_circ)
new_circ.svg()
```

```python
from algorithm.mapping_sampler import DifferentialMappingSearch
dsolver = DifferentialMappingSearch(circ, topo, 4, 3, no_extra=False)
new_circ, init_mapping, final_mapping = dsolver.solve(n_sample=5, n_iter=100, sabre_iter=1, lr=0.1)
print(f"initial mapping: {init_mapping}")
print(f"  final mapping: {final_mapping}")
print('gates in circuit:', len(circ))
print('  in new circuit:', len(new_circ))
new_circ.svg()
```
